```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 사용 가능한 폰트 리스트 출력
font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')

# 사용할 한글 폰트 경로 찾기 (예: 'NanumBarunGothic')
# 이 부분은 사용자의 시스템에 따라 달라질 수 있습니다.
# 아래 코드는 'Nanum'이라는 단어가 들어간 폰트를 찾습니다. 다른 폰트를 사용하려면 이름을 변경하세요.
korean_font = [f for f in font_list if 'Nanum' in f]
print(korean_font)  # 사용 가능한 한글 폰트 출력

# Matplotlib의 기본 폰트를 한글 폰트로 설정
plt.rcParams['font.family'] = 'NanumBarunGothic'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumBarunGothic'

# 데이터 불러오기
data = pd.read_csv('주요관광지점 입장객 2022년.csv')

# '총계' 열을 기준으로 내림차순 정렬
sorted_data = data.sort_values(by='총계', ascending=False)

# 막대 그래프 그리기
plt.figure(figsize=(10, 8))  # 그래프 크기 설정
plt.bar(sorted_data['관광지'], sorted_data['총계'])  # 막대 그래프 생성
plt.xlabel('관광지')  # x축 라벨
plt.ylabel('총계')  # y축 라벨
plt.title('2022년 관광지별 입장객 수')  # 그래프 제목
plt.xticks(rotation=90)  # x축 라벨 회전
plt.show()  # 그래프 출력
```
