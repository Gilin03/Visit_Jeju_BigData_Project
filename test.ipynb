```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import find_peaks

# 데이터 파일 불러오기
# 파일 경로를 데이터 파일의 실제 위치로 변경하십시오.
data = pd.read_csv('주요관광지점 입장객 2022년.csv')

# '총계' 열을 기준으로 내림차순 정렬
sorted_data = data.sort_values(by='총계', ascending=False)

# 변곡점 찾기
visitor_counts = np.array(sorted_data['총계'])
peaks, _ = find_peaks(visitor_counts, distance=1)

# 변곡점을 표시하기 위한 준비
peak_values = visitor_counts[peaks]
peak_names = sorted_data['관광지'].iloc[peaks]

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumBarunGothic'

# 막대 그래프와 함께 변곡점 표시
plt.figure(figsize=(10, 8))
plt.bar(sorted_data['관광지'], sorted_data['총계'], color='skyblue')
plt.scatter(peak_names, peak_values, color='red', s=100, zorder=2)  # 변곡점 표시
plt.xlabel('관광지')
plt.ylabel('총계')
plt.title('2022년 관광지별 입장객 수와 변곡점')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import folium

data = pd.read_csv('주요관광지점 입장객 2020~2022년_v1.csv')

# 위도와 경도를 가져옴
latitude = data.iloc[:, 4]
longitude = data.iloc[:, 5]
names = data.iloc[:, 2]

# 지도 생성 (초기 위치는 첫 번째 관광지의 위치로 설정)
m = folium.Map(location=[latitude[0], longitude[0]], zoom_start=10)

# 관광지 위치에 마커 추가
for lat, lon, name in zip(latitude, longitude, names):
    folium.Marker([lat, lon], popup=name).add_to(m)

# 지도 표시
m
```
